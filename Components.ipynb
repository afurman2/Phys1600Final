```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import libem
import numpy as np
from libvis import Visualizations
import matplotlib.pyplot as plt

from components import *
```

### Point Charge

```python
sim = libem.EMSimulationSpace3D(space_size=(2, 2, 2), top_left=(-1, -1, -1), scale=25)   
sim.compute(make_enforcer(
    enf(EMObjects.point_charge, (0, 0, 0), 1)
))
print("Done.")
```

```python
Visualizations.colormesh_3d(sim, size=(8, 8), resolution=1, color_norm=1)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.gca()
sim2d = libem.EMSimulationSpace2D.from_3d(sim, axis=0, location=0.5)
Visualizations.colormesh_2d(sim2d, graph_ax=ax)
Visualizations.efield_2d(sim2d, graph_ax=ax)
Visualizations.contour_2d(sim2d, graph_ax=ax)
plt.show()
```

### Rectangular Prism (Solid)

```python
sim = libem.EMSimulationSpace3D(space_size=(2, 2, 2), top_left=(-1, -1, -1), scale=25)   
sim.compute(make_enforcer(
    enf(EMObjects.rectangular_prism_solid, (-0.25, -0.5, -0.1), (0.5, 1, 0.2), 1)
))
print("Done.")
```

```python
Visualizations.colormesh_3d(sim, size=(8, 8), resolution=1, color_norm=1)
```

```python
Visualizations.color_xsections_3d(sim, 
                                  [(0, -0.5 + (0.25 * i)) for i in range(5)],
                                  size=(8, 8), resolution=1, color_norm=1)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.gca()
sim2d = libem.EMSimulationSpace2D.from_3d(sim, axis=0, location=0)
Visualizations.colormesh_2d(sim2d, graph_ax=ax)
Visualizations.efield_2d(sim2d, graph_ax=ax)
Visualizations.contour_2d(sim2d, graph_ax=ax)
plt.show()
```
